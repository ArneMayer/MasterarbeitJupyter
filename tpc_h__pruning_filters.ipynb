```python
import math

CHUNK_SIZE = 100_000

def create_statistics(df, chunk_size = CHUNK_SIZE):
    chunk_statistics = {}

    for chunk_id in range(math.ceil(len(df) / CHUNK_SIZE)):
        chunk_start = chunk_id * CHUNK_SIZE
        chunk_end = min((chunk_id+1) * CHUNK_SIZE, len(df))
#         print('== Chunk #{}: \t{} - {}'.format(chunk_id, chunk_start, chunk_end))
        df_chunk = df.iloc[chunk_start:chunk_end]
        chunk_statistics[chunk_id] = {'length': chunk_end - chunk_start}
        for attr in predicates:
#             print('{}: \t {} - {}'.format(attr, df_chunk[attr].min(), df_chunk[attr].max()))
            chunk_statistics[chunk_id][attr] = {'min': df_chunk[attr].min(),
                                                'max': df_chunk[attr].max()}
    return chunk_statistics
```

We build our exemplary data structure for each chunk

`{
    'chunk_id 1' : {
        length: ...,
        'stat_attr #1': {min: ..., max: ...}
    }    
...
}
`

```python
cs = create_statistics(lineitems_full)
# cs = create_statistics(lineitems_full_sorted)
```
